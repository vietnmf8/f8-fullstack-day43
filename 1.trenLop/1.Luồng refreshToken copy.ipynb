```python
# Hệ thống bảo mật
- Tưởng tượng: Khi bạn ra khỏi cửa mà không khoá nhà
  > Ai cũng có thể vào được
  -  Tức là dữ liệu không được bảo mật:
    > Ai cũng có thể tải được, truy cập, lấy được

- Cách bảo mật tài sản của chúng ta trong nhà
  > Ra khỏi nhà và khoá cửa
  > Ai có chìa thì mới có thể vào nhà của chúng ta
  
- `access_token`:
  + access: truy cập
  + token: chìa khoá
  > chính là "chìa khoá" để ra vào nhà chúng ta
  > có "chìa khoá" này là có vào nhà và lấy được "dữ liệu"

- Vấn đề:
  + Ở ngoài đời, chìa khoá thật không bao giờ hết hạn
    + Giả sử để chìa khoá linh tinh
    + Kẻ trộm có thể "clone" chìa khoá
    + Mặc dù chúng ta có "chìa khoá" nhưng người khác vẫn có thể vào nhà và lấy đồ được
  + `access_token` bị lộ cũng giống như việc chúng ta làm lộ chìa khoá
    + Không nhất thiết là bị mất "chìa khoá"
    + Bạn cầm "chìa khoá" đó, nhưng người khác cũng cầm
    > Không an toàn

- Giải pháp:
  + Ở ngoài đời: Cứ 1 tiếng, thay một cặp "chìa khoá - ổ khoá MỚI"
    > Chìa khoá cũ không thể mở được nhà nữa
  + Tư duy bảo mật của `access_token` cũng thế:
    > Tôi sợ lộ chìa
    > Tôi giảm rủi ro lộ chìa bằng cách một thời gian ngắn là tôi cứ thay một cặp "chìa khoá - ổ khoá MỚI"
    > Nếu "chìa khoá trước" bị lộ thì họ cũng không thể mở cửa nhà tôi vì tôi đã thay "ổ khoá mới" rồi!
  + Người ta thường để `access_token` có hạn sống ngắn (Khoảng 1 tiếng)

- Sự bất tiện:
  + Nhà của chúng ta là một ngôi nhà thông minh
  + Để đảm bảo tính bảo mật, cứ 1 tiếng nhà chúng ta đổi password một lần
  + Vậy mỗi lần mà chúng ta muốn quay về nhà để lấy đồ thì nhà đã bị thay password mất dồi
  > Vấn đề xảy ra: Chúng ta dùng chính password cũ thì lại không vào nhà được nữa
- Giải pháp:
  + Lúc đó chúng ta phải thực hiện xác thực lại
  + Tức là khai báo tên, mã số bí mật (mật khẩu)

# TÓM LẠI NHƯ SAU:
- Nhà bạn có một bác quản gia
- Khi muốn cho ai vào nhà, bác cần phải biết người đó là ai
- Bạn đăng ký danh tính (thông tin xác thực) với bác để bác nhận biết là bạn:
  + Cháu tên là... (username)
  + Căn cước của cháu là ... (password)
  > Sau khi bác quản gia nhận biết được bạn thông qua thông tin xác thực
  > Bác cấp cho bạn một "chìa khoá" đặc biệt (bác gọi là `access_token`)
  > Bác dặn: Lần sau cháu muốn vào nhà, chỉ cần đem theo `access_token`:
    + Bác sẽ mở cửa
    + Bác không cần biết cháu là ai (danh tính là gì)
    + Vì ai có `access_token` tức là đã được bác xác thực
  > Nhưng có trường hợp cháu vô tình để lộ `access_token`
    + Thì sẽ có người "clone" được `access_token`
    + Kẻ trộm lợi dụng và giơ `access_token` cho bác
    + Bác đơn giản chỉ nghĩ ai có "chìa khoá" này, bác sẽ mở cửa
    + Bác bị thủ tiêu, kẻ trộm đã lấy được đồ
  > Thế nên, CỨ 1 TIẾNG, bác thay đổi "chìa khoá đặc biệt" 1 lần
    + Trong khoảng 1 TIẾNG này, khi được cấp "chìa khoá"
    + Cháu có thể ra vào tự do, vì chìa khoá vẫn còn hiệu lực
    + Nhưng sau 1 TIẾNG cháu đi chơi về và muốn mở cửa:
      + Bác đã thay "chìa khoá mới"
      + Tuy nhiên cháu vẫn đang cầm "chìa cũ"
      + Nên giơ cho bác -> bác từ chối vì không còn phù hợp nữa
    > Mục tiêu: Bác giữ an toàn cho bác và cả gia đình cháu ạ
    + Lúc này bạn lại phải xác thực lại với quản gia rằng: 
      + Cháu tên là... (username)
      + Căn cước của cháu là ... (password)
    + Lúc này bác sẽ biết được cháu là chủ căn nhà này
    + Bác sẽ đưa cho cháu "chìa khoá đặc biệt" mới
    + Hạn chìa khoá này cũng 1 TIẾNG bác thay mới nhé
>> Giống cái việc là `access_token` hết hạn thì bạn phải đăng nhập lại như bình thường
>> Khá bất tiện: Cứ mỗi lần không có chìa, là lại phải gõ cửa xác thực lại danh tính, quá mất thời gian

# GIẢI PHÁP:
- Người thông minh như bạn có một deal với quản gia:
  + Vậy mỗi cháu xác thực và bác cấp cho cháu "chìa khoá access_token mới"
  + Thì bác cấp cho cháu một "Mã số bí mật" mà chỉ có 2 chúng ta biết mới với nhau
  + "Mã bí mật" này 3 tháng bác thay một lần
- Sau này nếu cháu về nhà mà "chìa khoá `access_token`" mà hết hạn:
  + Cháu không cần xác minh lại danh tính (cháu không cần phải đọc rõ họ tên, căn cước công dân)
  + Mà cháu chỉ cần đọc "mã bí mật"
  + Bác sẽ hiểu ra đây là cháu
  + Bác cấp cho cháu "chìa khoá `access_token` mới" kèm theo "mã bí mật" mới để đảm bảo an toàn tuyệt đối
  + Lúc này cháu có thể vào nhà
  >> Rút ngắn thời gian

- Trong lập trình web:
  + Khi đăng nhập (xác thực thông tin)
  + Server (bác quản gia) sẽ trả cho bạn:
    + `access_token` ("chìa khoá")
    + `refresh_token` ("mã bí mật")
```

```python
- **Bước 1**: Khi đăng nhập, gửi request API với thông tin đăng nhập 
  + Phía Authorization Server sẽ kiểm tra thông tin bạn gửi lên (email, password) có đúng không?
    > Sai: 401
- **Bước 2**: Nếu Đúng: Trả về response gồm: `access_token` + `refresh_token`
- **Bước 3**: Khi đó ở phía Client chúng ta đem theo `access_token` có quyền truy cập vào tài nguyên được bảo vệ (Resource Server)
  + Bình thường nếu không có `access_token`, chúng ta truy cập (gửi request) vào Resource Server sẽ bị lỗi 401
  + Chúng ta thực hiện request và đem theo `access_token`
    > Lưu ý: `access_token` là token để truy cập tài nguyên
    > dùng `refresh_token` không thể thay thế được
  + Resource Server đọc và kiểm tra `access_token` xem có hợp lệ không?
    > Không hợp lệ: trả về lỗi 401 Unauthorized
- **Bước 4**: Hợp lệ: Trả về tài nguyên được bảo vệ

- Khoảng một thời gian `access_token` hết hạn
- **Bước 5**: Tôi truy cập lại tài nguyên được bảo vệ (Resource Token)
- **Bước 6**: Token hết hạn --> trả về lỗi 401 (Invalid token Error)

> Bình thường đến tình huống này, chúng ta đá người dùng về trang đăng nhập
> Người dùng lại phải cung cấp lại thông tin --> Đây chính là sự bất tiện

- **Bước 7**: Refresh Token
  + Chúng ta không vội vàng đá người dùng về trang login nữa
  + Chúng ta lấy `refresh_token` từ lần gọi thành công
  + Chúng ta sẽ lưu lại ở `localStorage` hoặc trên cookie

  - Khi gặp lỗi 401 ở bước 6
  >> Chúng ta cần kiểm tra xem có `refresh_token` hay không?
  >  Nếu có , thực hiện request lên một API [POST] `refresh_token`
  > Lúc này, phía Authorization sẽ kiểm tra `refresh_token` có hợp , còn sống không?
- **Bước 8**: Nếu sống --> trả ra một cặp `access_token` và `refresh_token` mới
  + Khi đó chúng là dùng cặp `access_token` và `refresh_token` mới này để thử gọi lại request bị lỗi ở bước 5, bước 6
  + Chúng ta cũng lưu đè cặp `access_token` và `refresh_token` mới lên cặp cũ => Nhờ đó thời gian sống của chúng ta sẽ lâu dần lên
  + Đương nhiên phía Server khi trả ra cặp token MỚI -> sẽ HUỶ cặp token CŨ (kể ra token đó vẫn còn sống) để tăng tính bảo mật
  + VD: ở bước 7 gửi request với `refresh_token` còn sống
  + Nhưng khi ở bước 8 trả về `refresh_token` mới thì nó huỷ `refresh_token` cũ ở bước 7 đi
```

```python
# Thực tế: 
- Một trang web thực thụ, Khi ở một trang có thể gọi cùng lúc nhiều API cần được bảo vệ
- Vậy nếu ở bước 5: Nếu gọi cùng lúc 10 cái API thì sẽ như thế nào?
  + Bây giờ token của mình hết hạn rồi, nếu bắn 10 API như vậy
  + Thì ở bước 6 sẽ lỗi 10 lần!!
  + Và rất có ở thể bước 7 sẽ bắn 10 cái `refresh_token` đi
? Vậy điều gì sẽ xảy ra khi bắn 10 cái `refresh_token` đi
  > Hãy nhớ rằng: khi Server trả ra cặp token MỚI -> sẽ HUỶ cặp token CŨ (kể ra token đó vẫn còn sống)
  > Tức là nếu bắn 10 request cái với `refresh_token` (CŨ): 
    + Cái nào đến trước (VD API 1 đến trước ở bước 7)
    + Thành công: Trả ra cặp token MỚI --> xoá token (CŨ)
  > Tuy nhiên API 2 -> 10 request với `refresh_token` CŨ -> nhưng nó đã bị HUỶ sau khi API 1 thành công
    + Server Authorization kiểm tra `refresh_token` đã chết rồi, không còn hạn nữa => lỗi  

# NÔM NA ví dụ quản gia
- Giống như cả nhà đi làm về lần lượt là CON -> BỐ -> MẸ và cùng chuẩn bị vào nhà
- CON - BỐ - MẸ đều có "mã bí mật" (`refresh_token`) để không cần khai báo đầy đủ với quản gia mà vẫn có thể vào được nhà
- vì CON nhanh chân nên chui vào và giơ "chìa khoá lên trước"
  + Bối cảnh: Sau khi "chìa khoá `access_token` hết hạn
  + CON về nhà trước là giơ "mã bí mật" (`refresh_token`) cho quản gia
  + Quản gia kiểm tra "mã bí mật" hợp lệ và cấp cho CON "chìa khoá `access_token` và mã bí mật `refresh_token` mới và ông quản gia HUỶ "mã bí mật" CŨ đi
  + Tuy nhiên, BỐ, MẸ về sau vẫn cầm "mã bí mật CŨ" để đòi ông quản gia mở cửa
  + Ông quản gia kiểm tra "mã bí mật" của Bố , Mẹ đã cũ, tôi huỷ nó rồi, nó không còn hợp lệ nữa
  + BỐ, MẸ ngủ ngoài đường!!

>> ĐÓ CHÍNH LÀ VẤN ĐỀ
# GIẢI PHÁP
  1. Thay vì để CON - BỐ - MẸ cùng vào nhà cùng lúc
  2. Thì chỉ có CON vào nhà và nhận cặp "chìa + mã bí mật" mới từ quản gia
  3. BỐ và MẸ ngồi im, ngồi chờ CON cầm "chìa + mã bí mật" phát cho BỐ và MẸ
  4. BỐ và MẸ có cặp "chìa + mã bí mật" mới --> có thể vào nhà
```
